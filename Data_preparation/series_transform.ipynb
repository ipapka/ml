Develop a Python function to automatically transform a series into a supervised learning
problem where the number of input and output time steps can be specified as arguments.

```python
import pandas as pd
df=pd.DataFrame()
# define example
df['t'] = [x for x in range(100)]
```

```python
# transform function
def func(input_,output_,df):
    for i in range(input_):
        name='input_'+str(i)
        df[name]=df['t'].shift(output_+i)
    for i in range(output_):
        name='output_'+str(i)
        df[name]=df['t'].shift(i)
    df=df.drop('t', axis=1)
    return df   
```

```python
func(6,5,df)
```

'input' columns are the input data for the supervised learning problem and 'output' - are the targets
